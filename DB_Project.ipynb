```python
%%sql
-- ALTER TABLE Theaters DROP CONSTRAINT LocationID;
-- Drop the "Locations" table
DROP TABLE IF EXISTS Locations CASCADE;
DROP TABLE IF EXISTS Movies CASCADE;
DROP TABLE IF EXISTS Theaters CASCADE;
DROP TABLE IF EXISTS Showtimes CASCADE;
DROP TABLE IF EXISTS Users CASCADE;
DROP TABLE IF EXISTS Seats CASCADE;
DROP TABLE IF EXISTS Bookings CASCADE;
DROP TABLE IF EXISTS Payments CASCADE;
DROP TABLE IF EXISTS Coupons CASCADE;
-- Create Locations table
CREATE TABLE Locations (
    LocationID SERIAL PRIMARY KEY,
    Address VARCHAR(255),
    City VARCHAR(100),
    State VARCHAR(100),
    ZipCode VARCHAR(20)
);

-- Create Theaters table
CREATE TABLE Theaters (
    TheaterID SERIAL PRIMARY KEY,
    Name VARCHAR(255),
    LocationID INT,
    NumberOfScreens INT,
    FOREIGN KEY (LocationID) REFERENCES Locations(LocationID)
);
CREATE TABLE Seats (
    SeatID SERIAL PRIMARY KEY,
    TheaterID INT,
    RowNumber INT,
    SeatNumber INT,
    IsAvailable BOOLEAN DEFAULT TRUE,
    FOREIGN KEY (TheaterID) REFERENCES Theaters(TheaterID)
);

-- Create Movies table
CREATE TABLE Movies (
    MovieID SERIAL PRIMARY KEY,
    Title VARCHAR(255),
    Duration INT, -- Duration in minutes
    Genre VARCHAR(50)
);

-- Create Showtimes table
CREATE TABLE Showtimes (
    ShowtimeID SERIAL PRIMARY KEY,
    MovieID INT,
    SeatID INT,
    Showtime TIMESTAMP,
    FOREIGN KEY (MovieID) REFERENCES Movies(MovieID),
    FOREIGN KEY (SeatID) REFERENCES Seats(SeatID)
);


-- Create Users table
CREATE TABLE Users (
    UserID SERIAL PRIMARY KEY,
    FullName VARCHAR(255),
    Email VARCHAR(255),
    Password VARCHAR(255) -- Consider hashing in a real application
);

-- Create Bookings table
CREATE TABLE Bookings (
    BookingID SERIAL PRIMARY KEY,
    UserID INT,
    ShowtimeID INT,
    NumberOfTickets INT,
    BookingTime TIMESTAMP DEFAULT NOW(),
    FOREIGN KEY (UserID) REFERENCES Users(UserID),
    FOREIGN KEY (ShowtimeID) REFERENCES Showtimes(ShowtimeID)
);

-- Create Payments table
CREATE TABLE Payments (
    PaymentID SERIAL PRIMARY KEY,
    BookingID INT,
    Amount DECIMAL(10, 2),
    PaymentMethod VARCHAR(50),
    PaymentTime TIMESTAMP DEFAULT NOW(),
    FOREIGN KEY (BookingID) REFERENCES Bookings(BookingID)
);

-- Create Coupons table
CREATE TABLE Coupons (
    CouponID SERIAL PRIMARY KEY,
    Code VARCHAR(50),
    DiscountAmount DECIMAL(10, 2),
    ExpiryDate DATE,
    UserID INT
);

-- Add a CouponID column to the Bookings table to link a booking with a coupon (optional relationship)
ALTER TABLE Bookings ADD COLUMN CouponID INT;
ALTER TABLE Bookings ADD FOREIGN KEY (CouponID) REFERENCES Coupons(CouponID);

-- Statements to populate the tables with dummy data will go here
-- Example:
-- INSERT INTO Locations (Address, City, State, ZipCode) VALUES ('1234 Movie St', 'Filmville', 'CA', '90001');
-- Repeat for other tables as needed
```

```python
%%sql
DROP FUNCTION IF EXISTS CheckSeatAvailability(INT, INT);
CREATE OR REPLACE FUNCTION CheckSeatAvailability(ShowtimeID INT, SeatID INT) RETURNS INT AS $$
DECLARE
    TotalSeats INT;
    BookedSeats INT;
    RemainingSeats INT;
BEGIN
    SELECT COUNT(*) INTO TotalSeats
    FROM Seats;
    SELECT SUM(NumberOfTickets) INTO BookedSeats
    FROM Bookings
    WHERE Bookings.ShowtimeID = CheckSeatAvailability.ShowtimeID;

    -- Calculate remaining seats
    RemainingSeats := TotalSeats - COALESCE(BookedSeats, 0);

    RETURN RemainingSeats;
END;
$$ LANGUAGE plpgsql;



```

```python
%%sql
SELECT CheckSeatAvailability(1, 2) AS RemainingSeats;
```

```python
%%sql
DROP FUNCTION IF EXISTS GetUsersWatchingMovie(INT);
CREATE OR REPLACE FUNCTION GetUsersWatchingMovie(MovieID INT) RETURNS TABLE (
    MovieName VARCHAR(255),
    UserID INT,
    FullName VARCHAR(255),
    Email VARCHAR(255)
) AS $$
BEGIN
    RETURN QUERY
    SELECT DISTINCT M.Title AS MovieName, U.UserID, U.FullName, U.Email
    FROM Users U
    INNER JOIN Bookings B ON U.UserID = B.UserID
    INNER JOIN Showtimes S ON B.ShowtimeID = S.ShowtimeID
    INNER JOIN Movies M ON S.MovieID = M.MovieID
    WHERE M.MovieID = GetUsersWatchingMovie.MovieID;
END;
$$ LANGUAGE plpgsql;

```

```python
%%sql
CREATE OR REPLACE FUNCTION GetUserCouponPayments()
RETURNS TABLE (
    UserID INT,
    FullName VARCHAR(255),
    CouponCode VARCHAR(50),
    PaymentMethod VARCHAR(50)
) AS $$
BEGIN
    RETURN QUERY
    SELECT U.UserID, U.FullName, C.Code AS CouponCode, P.PaymentMethod
    FROM Users U
    JOIN Bookings B ON U.UserID = B.UserID
    LEFT JOIN Payments P ON B.BookingID = P.BookingID
    LEFT JOIN Coupons C ON B.CouponID = C.CouponID;
END;
$$ LANGUAGE plpgsql;
```

```python
%%sql
SELECT * FROM GetUserCouponPayments();
```

```python
%%sql
CREATE OR REPLACE FUNCTION CalculateTotalIncome() RETURNS DECIMAL AS $$
DECLARE
    TotalIncome DECIMAL;
BEGIN
    SELECT SUM(Amount) INTO TotalIncome
    FROM Payments;

    RETURN TotalIncome;
END;
$$ LANGUAGE plpgsql;
```

```python
%%sql
SELECT CalculateTotalIncome() AS TotalIncome;
```
